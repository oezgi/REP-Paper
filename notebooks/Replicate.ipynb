## Changing Statistical Tests for Assessing Differences Between Home and Away Matches

### Using Paired t-Test instead of Wilcoxon 

```python
results_stat = pd.DataFrame()
```

### Ligue 1

```python
dataframes = {
    2014: df_ligue1_2014,
    2015: df_ligue1_2015,
    2016: df_ligue1_2016,
    2017: df_ligue1_2017,
    2018: df_ligue1_2018,
    2019: df_ligue1_2019,
    2020: df_ligue1_2020,
}
```

```python
for season, df in dataframes.items():
    # Filter data for home and away games
    home_pts = df[df['h_a'] == 'h']['pts']
    away_pts = df[df['h_a'] == 'a']['pts']
    home_xpts = df[df['h_a'] == 'h']['xpts']
    away_xpts = df[df['h_a'] == 'a']['xpts']
    home_xg = df[df['h_a'] == 'h']['xG']
    away_xg = df[df['h_a'] == 'a']['xG']
    
    # Perform Paired t-Test
    ttest_result = stats.ttest_rel(home_pts, away_pts)
    ttest_xpts = stats.ttest_rel(home_xpts, away_xpts)
    ttest_xg = stats.ttest_rel(home_xg, away_xg)
    
    # Store results in a dictionary
    result = {
        "league": "Ligue_1",
        "season": season,
        "ttest_result": ttest_result.statistic,
        "ttest_result-pvalue": ttest_result.pvalue,
        "ttest-xPoints": ttest_xpts.statistic,
        "ttest-xPoints-pvalue": ttest_xpts.pvalue,
        "ttest-xG": ttest_xg.statistic,
        "ttest-xG-pvalue": ttest_xg.pvalue,
    }
    
    # Append the result to the results list
    results_stat = pd.concat([results_stat, pd.DataFrame([result])], ignore_index=True)
```

## Post-COVID Seasons as a Variability Factor

In this replicability analysis, we aim to assess the impact of the post-COVID seasons (2021 onward) as a potential variability factor in understanding home advantage in football. The COVID-19 pandemic created a unique natural experiment in sports, where matches were played in empty stadiums, effectively removing one of the critical elements of home advantage: crowd support. This provided a rare opportunity to isolate the influence of environmental and psychological factors on team performance.

However, with the return of fans to stadiums in subsequent seasons, an important question arises: **Did home-field advantage return to pre-COVID levels, or has the dynamic shifted permanently?** This variability factorperformance in post-COVID seasonswill allow us to test the robustness and generalizability of the original study's conclusions.

### Statistical Analysis of Home Advantage

By performing **Wilcoxon Signed-Rank** tests and calculating **Cohens d effect** sizes, we aim to quantify the differences between home and away performances both before, during, and after the pandemic.
This analysis investigates:
1. **Statistical Significance:** Using Wilcoxon p-values to determine whether differences in metrics such as points (pts), expected points (xPoints), and expected goals (xG) between home and away teams are significant.
2. **Effect Sizes:** Using Cohens d to measure the magnitude of these differences and assess the practical impact of home advantage.

#### Ligue 1

```python
dataframes = {
    # Ligue 1
    2014: df_ligue1_2014,
    2015: df_ligue1_2015,
    2016: df_ligue1_2016,
    2017: df_ligue1_2017,
    2018: df_ligue1_2018,
    2019: df_ligue1_2019,
    2020: df_ligue1_2020,
    2021: df_ligue1_2021,
    2022: df_ligue1_2022,
    2023: df_ligue1_2023,
}
```

```python
for season, df in dataframes.items():
    # Filter data for home and away games
    home_pts = df[df['h_a'] == 'h']['pts']
    away_pts = df[df['h_a'] == 'a']['pts']
    home_xpts = df[df['h_a'] == 'h']['xpts']
    away_xpts = df[df['h_a'] == 'a']['xpts']
    home_xg = df[df['h_a'] == 'h']['xG']
    away_xg = df[df['h_a'] == 'a']['xG']
    
    # Perform Wilcoxon Signed-Rank tests
    wilco_result = stats.wilcoxon(home_pts, away_pts, alternative='two-sided')
    wilco_xpts = stats.wilcoxon(home_xpts, away_xpts, alternative='two-sided')
    wilco_xg = stats.wilcoxon(home_xg, away_xg, alternative='two-sided')
    
    # Calculate Cohen's d for each metric
    cohend_result = (home_pts.mean() - away_pts.mean()) / np.sqrt((home_pts.std() ** 2 + away_pts.std() ** 2) / 2)
    cohend_xpts = (home_xpts.mean() - away_xpts.mean()) / np.sqrt((home_xpts.std() ** 2 + away_xpts.std() ** 2) / 2)
    cohend_xg = (home_xg.mean() - away_xg.mean()) / np.sqrt((home_xg.std() ** 2 + away_xg.std() ** 2) / 2)
    
    # Store results in a dictionary
    result = {
        "league": "Ligue_1",
        "season": season,
        "wilco-result": wilco_result.statistic,
        "wilco-result-pvalue": wilco_result.pvalue,
        "result-cohend": cohend_result,
        "wilco-xPoints": wilco_xpts.statistic,
        "wilco-xPoints-pvalue": wilco_xpts.pvalue,
        "xPoints-cohend": cohend_xpts,
        "wilco-xG": wilco_xg.statistic,
        "wilco-xG-pvalue": wilco_xg.pvalue,
        "xG-cohend": cohend_xg,
    }
    
    # Append the result to the results list
    results_stat = pd.concat([results_stat, pd.DataFrame([result])], ignore_index=True)
```

```python
# Function to extract the season based on the date string
def assign_season_from_string(date_str):
    if pd.isna(date_str):
        return None  # Handle missing dates explicitly
    # Extract year and month from the date string
    year = int(date_str[:4])  # First 4 characters represent the year
    month = int(date_str[5:7])  # Characters 5-7 represent the month
    if month >= 9:  # Matches between September and December belong to the current year
        return year
    elif month <= 8:  # Matches between January and August belong to the previous year
        return year - 1
    else:
        return None  # This should not occur
```

```python
# Function to clean and transform the DataFrame
def new_dataframe(df): 
    # Rename the 'Team' column to 'team_name'
    df.rename(columns={"Team": "team_name"}, inplace=True)
    df.rename(columns={"Round": "match_round"}, inplace=True)
    
    # Update 'h_a' column logic
    df['h_a'] = df['h_a'].apply(lambda x: 'h' if pd.isna(x) else 'a')

    # Extract the season from the 'Date' column (kept as string)
    df['season'] = df['Date'].apply(assign_season_from_string)
    df = df[df['season'].notna()]  # Filter out rows with missing seasons
    df['season'] = df['season'].astype(int)  # Ensure 'season' is an integer

    # Keep only the necessary columns
    columns_to_keep = ['team_name', 'season', 'h_a', 'xG', 'match_round', 'GF', 'GA', 'GD']
    df = df[columns_to_keep]
    
    return df
```

### Table with bar charts to visualize and investigate a possible effect of COVID on home advantage

```python
# Function to calculate grouped results
def calculate_seasonal_sums(df):
    # Group by season and calculate the sums
    grouped = df.groupby('season').apply(lambda group: pd.Series({
        'sum_points_away': group.loc[group['h_a'] == 'a', 'GF'].sum(),  # Sum of GF where h_a == 'a'
        'sum_points_home': group.loc[group['h_a'] == 'h', 'GF'].sum(),  # Sum of GF where h_a == 'h'
        'sum_xG_away': group.loc[group['h_a'] == 'a', 'xG'].sum(),      # Sum of xG where h_a == 'a'
        'sum_xG_home': group.loc[group['h_a'] == 'h', 'xG'].sum(),      # Sum of xG where h_a == 'h'
        'avg_points_away':  group.loc[group['h_a'] == 'a', 'GF'].mean(),
        'avg_points_home':  group.loc[group['h_a'] == 'h', 'GF'].mean(),
        'avg_xG_away': group.loc[group['h_a'] == 'a', 'xG'].mean(),      # Sum of xG where h_a == 'a'
        'avg_xG_home': group.loc[group['h_a'] == 'h', 'xG'].mean()
    })).reset_index()

    return grouped
```

```python
results = calculate_seasonal_sums(df_ucl)
```

```python
diff_results = pd.DataFrame({
    "LEAGUE": 'Champions League',
    "YEAR": results["season"],      # Keep the year column
    "DIFF_POINTS_HOMEAWAY": (results["sum_points_home"] - results["sum_points_away"]).round(0).astype(int),  # Calculate points difference
})
```

```python
styled_df = diff_results.style.bar(
    subset=["DIFF_POINTS_HOMEAWAY"], 
    color=['lightcoral', 'mediumseagreen'], 
    align='zero'  # Aligns negative bars to the left
)
styled_df
```

We can put forward the hypothesis that the reason for this is the absence of crowds. Home advantage is often considered more pronounced in international tournaments due to long-distance travel, unfamiliar environments, and crowd hostility.
